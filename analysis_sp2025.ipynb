```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/sp2025.csv')
df.head()
```

```python
df = df.drop(columns=['Term', 'A Range', 'B Range', 'C Range', 'D Range'])
df.head()
```

```python
num_cols = ['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D', 'D-', 'F', 'Students']
agg_dict = {col: 'sum' for col in num_cols}
agg_dict.update({
    'Course Subject': 'first',
    'Course Number': 'first'
})

df['Students'] = df['A+'] + df['A'] + df['A-'] + df['B+'] + df['B'] + df['B-'] + df['C+'] + df['C'] + df['C-'] + df['D+'] + df['D'] + df['D-'] + df['F']
df['Course'] = df['Course Subject'] + ' ' + df['Course Number'].astype(str)
df_grouped = df.groupby('Course').agg(agg_dict).reset_index()
df_grouped.head()
```

```python
df_grouped['Average Grade'] = (4.0 * df_grouped['A+'] + 4.0 * df_grouped['A'] + 3.7 * df_grouped['A-'] + 3.3 * df_grouped['B+'] + 3.0 * df_grouped['B'] + 2.7 * df_grouped['B-'] + 2.3 * df_grouped['C+'] + 2.0 * df_grouped['C'] + 1.7 * df_grouped['C-'] + 1.3 * df_grouped['D+'] + 1.0 * df_grouped['D'] + 0.7 * df_grouped['D-'] + 0.0 * df_grouped['F']) / df_grouped['Students']
df_grouped.head()
```
